# Introduction

In Assignment 10 for our class, we looked at the use of Q-Learning (QL) and Deep Reinforcement Learning (DRL) through application to Blackjack (BJ). A Medium article, entitled "Blackjack, Stocks, and Reinforcement Learning: ML for Risk Management" [1, 2] ("the Article") was provided for us to review.  This article and the assignment, while useful for learning, left certain issues unaddressed and the purpose of this project is to better evaluate these algorithms as applied to BJ.

The following issues with the article were noted:
1. The results obtained seem unrealistic. For QL, the author demonstrated a reward per episode of -68/100, or -68%. For DRL, the author demonstrated an expected return of 643%/100 episode, or +6.43% per episode. Blackjack is a well studied game that is believed to have an optimal strategy and these values are not consistent with that prior analysis.  Without counting cards or other cheating methods, optimally played BJ has a player expected value of about -0.5% [3], which implies a house advantage.  Even by counting cards or other cheats,  player EV can only be brought to +1.0% [4].  This implies that the results presented in our assignment may have been erroneous.
2. The Open AI Gym BJ environment has some shortcomings for teaching RL.  First, the output of the environment does not reveal player or dealer cards, making understanding the workings of the model challenging.  Second, the environment only offers a simplified game of BlackJack, which does not have Splits, Doubling Down, or Surrenders incorporated.  This simplified model is helpful for implementing RL tests but the missing options reduce the player EV.
3. The assigned article output is challenging to interpret for a few reasons.  First the many images of BJ games are hard to understand and dominate the output.  Second, the author does not compare his algorithms to any baseline, so performance is difficult to interpret.


To address these concerns, I decided to provide the following:
1. Reassessment of QL and DRL in the gym BJ environment, including:
    1. Establish baseline performance of a naive and expert strategies.
    2. Compare performance of the Article QL and DRL models to these baselines.
    3. Create new QL and DRL models to see if this performance can be improved.
2. Provide an additional implementation of BJ, which allows for Double Down, Surrender, and Split actions
    1. Assess performance of naive and expert models in this environment
    2. Train and assess new QL and DRL models in this environment.

Unfortunately, troubleshooting and training in the gym environment left me with insufficient time to move into a more complex environment adn I will have to pursue this beyond the class.

Hypotheses:
1. The results presented in the Medium article are incorrect.
2. The results presented in the Medium article can be improved upon with better models.
3. For blackjack, QL models will perform better than DRL models
4. AI models can approach but not exceed the performance of expert models in Blackjack.

# Notebook Setup

```python
import warnings
warnings.filterwarnings(action ='ignore', category=DeprecationWarning)

import pandas as pd

import gym
import strategies

import gym_utils
env = gym.make('Blackjack-v1')
```


# Analysis in the Gym environment

First, an assessment is provided of naive strategies and expert strategies. This is followed by assessment of the Medium article strategies and then the models developed for this project. Each strategy was addressed based on two criteria:
- Policy Score - The percentage of actions in a given strategy that agree with the expert strategy.  A higher percentage represents a better model.
- Expected Return - This is the average reward per hand after playing 5000 hands of blackjack with the strategy being tested.  5000 hands was chosen as there was substantial variability in result with fewer hands.  The standard deviation of results with 5000 hands was about 1%, which can be observed in the multiple tables presented.

## Naive and Expert Strategy Assessment

We begin by assessing performance of the naive (random action choice) and expert models.  This baseline is important for further analysis as it represents the worst case model and barrier above which we do not expect to perform.  In addition to the random strategy, I also assessed a model where 'Hit' is chosen every time and another where 'Stand' is chosen every time.  The table below show the performance of these models from best to worst.   It is no surprise that the 'all hit' model loses every game, so it is discarded from further analysis.  The 'all stand' policy performs better than the random policy, which provides two baselines for pour performance.

The expert strategy performs best but does not perform as well as the literature reports due to the absence of double, surrender, and split actions. Additionally, the gym environment defaults to a 1:1 payout for blackjack as opposed to the usual 3:2 payout.

```python
models = {'naive': 'random',
          'all_stand': strategies.ALL_STAND_STRATEGY,
          'all_hit': strategies.ALL_HIT_STRATEGY,
          'expert': strategies.BASIC_STRATEGY_HIT_STAND_ONLY}
df, results = gym_utils.evaluate_strategies(env, models)
df
```

```python
# Plot results over Time
gym_utils.plot_results(results, df.index)
```

```python
# Display Expert Strategy
expert_strategy = strategies.BASIC_STRATEGY_HIT_STAND_ONLY
expert_strategy
```

# Findings

#### Hypothesis 1 - The results presented in the Medium article are incorrect.

The results presented in the Medium article for Assignment 10 were invalid as shown by the retesting of the results herein.  The reported results for Q-Learning were worse than a naive agent (-68% vs. -40%) and those for DRL showed a marked positive return which is theoretically impossible.  Additionally, these results were not reproducible in my experiments.  There are a few potential causes for these results:
    - Q-Learning - I believe the error in the Medium article code was that the author did not set the Q-values for terminal states to zero.  This is a requirement of Q-Learning algorithms.  I made this error myself at first, and it led to unstable growth of the Q-values in model, once it was fixed the model converged nicely.
    - DRL - The main issue I found here is that the author used the same seed value for training and testing and I believe this led to overfitting of the model to this specific pseudorandom series of cards.

#### Hypothesis 2 - The results presented in the Medium article can be improved upon with better models.

The final figure above shows that the models developed for this project were superior to those in the article, thereby confirming the hypothesis.  I believe the reasons for this are correction of the issues mention in the section on Hypothesis 1 as well as improved tuning of the models.

#### Hypothesis 3 - For blackjack, QL models will perform better than DRL models

This hypothesis was not confirmed as the new QL and DRL models performed similarly to one another.  In hindsight, this makes sense as a well-developed DRL model should be able to simulate a Q table given sufficient memory size and iterations.

#### Hypothesis 4 - AI models can approach but not exceed the performance of expert models in Blackjack.

The evidence presented supports this hypothesis as the new models were similar but inferior to the expert in all cases.  Proof of this hypothesis would require an exhaustive search which is not possible to perform.

# Lessons learned
1. Converting data to human interpretable format such as the strategy maps in this discussion is a very helpful technique for evaluating the validity of a model as well as troubleshoot models.
2. One must be careful with blog type resources such as this Medium article.  They can be very useful for getting another take on a complex topic but the lack of peer review means that they may contain erroneous conclusion and/or invalid techniques. It is an assumption, however, using articles with a  greater number of followers is less likely to be misleading as the odds of the author having received feedback goes up.
3. Q-Learning and DRL are powerful tools for decision-making in the face of uncertainty.  Blackjack is a good model to start to evaluate this technique as the optimal policy is known for comparison.
4. As with most models, details are very important for making them work properly.  It is best to start small and build up as opposed to the opposite.  I started this project by trying to make the entire game of Blackjack work and I believe I would have gotten more done if I had started with the Gym environment and then advanced to a more complete environment.

# Future Work
I had planned on extending this work to a Blackjack game that incorporated doubles, surrenders, and splits.  In fact, I had even generated a substantial amount of code to implement this but ran short on time to complete the models.  A more complete Blackjack game carries a few advantages for learning more about QL and DRL:
1. This more complex game has been better studied and the expert policies are therefore more accurate.
2. The larger action space makes the model more complex and requires some additional code to interpret.
3. The action space varies with different states which adds in an interesting challenge.  For example, one can only split or double when one has 2 cards.  This can be handled by expanding the state space model but there is a better way.  BY changing the action space when searching for Qmax(s,a), a model can select the largest Q-value for available actions as opposed to the argmax.  This means a well-developed q table for the entire state action space can be used for limited action spaces as well.  It would be informative to put this into practice as other authors have stated DRL does not work in this situation. [8]
