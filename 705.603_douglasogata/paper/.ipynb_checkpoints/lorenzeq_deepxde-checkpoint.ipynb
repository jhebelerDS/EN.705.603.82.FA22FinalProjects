Define the Lorenz equations
$$
\begin{aligned}
\frac{dx}{dt} &= \sigma \left( y - x \right) \\
\frac{dy}{dt} &= x \left( \rho - z \right) - y \\
\frac{dz}{dt} &= xy - \beta z
\end{aligned}
$$

```python
import deepxde as dde
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import integrate
from deepxde.backend import tf
```

```python
sigma = 10.
rho = 22.
beta = 8./3.
```

```python
geom = dde.geometry.TimeDomain(0.,20.)
```

```python
def lorenz_system(t,Y):
    x = Y[:, 0:1]
    y = Y[:, 1:2]
    z = Y[:, 2:3]
    
    dxdt = dde.grad.jacobian(Y,t,i=0)
    dydt = dde.grad.jacobian(Y,t,i=1)
    dzdt = dde.grad.jacobian(Y,t,i=2)
    return [
        dxdt - sigma*(y - x),
        dydt - (x*(rho - z) - y),
        dzdt - (x*y - beta*z)
    ]
```

```python
data = dde.data.PDE(geom, lorenz_system, [], 400, 2, num_test=3000)
```

```python
layer_size = [1] + [40] * 3 + [3]
activation = "tanh"
initializer = "Glorot normal"
net = dde.nn.FNN(layer_size, activation, initializer)
```

Since we expect to have periodic behavior in the Lorenz equation, we add a feature layer with $sin(kt)$. This forces the prediction to be periodic and therefore more accurate.

```python
def input_transform(t):
    return tf.concat(
        (
            t,
            tf.sin(t),
            tf.sin(2 * t),
            tf.sin(3 * t),
            tf.sin(4 * t),
            tf.sin(5 * t),
            tf.sin(6 * t),
        ),
        axis=1,
    )
```

Specify initial conditions
`y0 = [7; 7; 5]`

```python
Y_init = [7, 7, 5]
```

```python
def boundary(_, on_initial):
    return on_initial
```

```python
ic1 = dde.icbc.IC(geom, lambda X: Y_init[0], boundary, component=0)
ic2 = dde.icbc.IC(geom, lambda X: Y_init[1], boundary, component=1)
ic3 = dde.icbc.IC(geom, lambda X: Y_init[2], boundary, component=2)
```

```python
data = dde.data.PDE(geom, lorenz_system, [ic1, ic2, ic3], 400, 2, num_test=50)
```

```python
def output_transform(t, Y):
    y1 = Y[:, 0:1]
    y2 = Y[:, 1:2]
    y3 = Y[:, 2:3]

    return tf.concat(
        [y1 * tf.tanh(t) + Y_init[0], y2 * tf.tanh(t) + Y_init[1], y3*tf.tanh(t)+Y_init[2]], axis=1
    )
```

```python
model = dde.Model(data, net)
model.compile("adam", lr=0.001)
losshistory, train_state = model.train(iterations=10000)
```

Continue training with `L-BFGS` to achieve smaller losses.

Define numerical integrator for the Lorenz system.

```python
def lorenzeq(t, Y):
    x, y, z = Y
    dxdt = sigma*(y - x)
    dydt = x*(rho - z) - y
    dzdt = x*y - beta*z
    return dxdt, dydt, dzdt
```

```python
def lorenzeq_jac(t, Y):
    x, y, z = Y
    J = np.array(
        [[-sigma,sigma,0],
        [rho, -1, -x],
        [y,x,-beta]]
    )
    return J
```

```python
def integrate_forward(func,t_range,Y_init,N,J):
    t = np.linspace(t_range[0], t_range[1], N)

    sol = integrate.solve_ivp(func, t_range, Y_init, method='LSODA',t_eval=t, jac=J)
    x,y,z = sol.y
    x = x.reshape(N, 1)
    y = y.reshape(N, 1)
    z = z.reshape(N, 1)
    
    return x,y,z
```

Calculate the trajectory numerically.

```python
N = 1000
t_range = (0.,20.)
ti,tf = t_range
t = np.linspace(ti, tf, N)
x_true, y_true, z_true = integrate_forward(lorenzeq,t_range,Y_init,N,lorenzeq_jac)
```

Calculate prediction.

```python
t = t.reshape(N, 1)
sol_pred = model.predict(t)
x_pred = sol_pred[:, 0:1]
y_pred = sol_pred[:, 1:2]
z_pred = sol_pred[:, 2:3]
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, x_true, color="black", label="x_true")
plt.plot(t, y_true, color="blue", label="y_true")
plt.plot(t, z_true, color="purple", label="z_true")

plt.plot(t, x_pred, color="red", linestyle="dashed", label="x_pred")
#plt.plot(t, y_pred, color="orange", linestyle="dashed", label="y_pred")
plt.plot(t, z_pred, color="cyan", linestyle="dashed", label="z_pred")


plt.xlabel("t")
plt.ylabel("Y")
plt.legend()
plt.show()
```
