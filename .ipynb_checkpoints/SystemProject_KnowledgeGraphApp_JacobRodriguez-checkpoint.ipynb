## Define functions

### Function to split text in document into individual sentences

```python
# -*- coding: utf-8 -*-
alphabets= "([A-Za-z])"
prefixes = "(Mr|St|Mrs|Ms|Dr)[.]"
suffixes = "(Inc|Ltd|Jr|Sr|Co)"
starters = "(Mr|Mrs|Ms|Dr|He\s|She\s|It\s|They\s|Their\s|Our\s|We\s|But\s|However\s|That\s|This\s|Wherever)"
acronyms = "([A-Z][.][A-Z][.](?:[A-Z][.])?)"
websites = "[.](com|net|org|io|gov)"
digits = "([0-9])"

def split_into_sentences(text):
    """
    Splits a document with text into individual sentences

    :param text (str): single string of text from document
    :return sentences (list): list of sentences from document
    """
    
    text = " " + text + "  "
    text = text.replace("\n"," ")
    text = re.sub(prefixes,"\\1<prd>",text)
    text = re.sub(websites,"<prd>\\1",text)
    text = re.sub(digits + "[.]" + digits,"\\1<prd>\\2",text)
    if "..." in text: text = text.replace("...","<prd><prd><prd>")
    if "Ph.D" in text: text = text.replace("Ph.D.","Ph<prd>D<prd>")
    text = re.sub("\s" + alphabets + "[.] "," \\1<prd> ",text)
    text = re.sub(acronyms+" "+starters,"\\1<stop> \\2",text)
    text = re.sub(alphabets + "[.]" + alphabets + "[.]" + alphabets + "[.]","\\1<prd>\\2<prd>\\3<prd>",text)
    text = re.sub(alphabets + "[.]" + alphabets + "[.]","\\1<prd>\\2<prd>",text)
    text = re.sub(" "+suffixes+"[.] "+starters," \\1<stop> \\2",text)
    text = re.sub(" "+suffixes+"[.]"," \\1<prd>",text)
    text = re.sub(" " + alphabets + "[.]"," \\1<prd>",text)
    if "”" in text: text = text.replace(".”","”.")
    if "\"" in text: text = text.replace(".\"","\".")
    if "!" in text: text = text.replace("!\"","\"!")
    if "?" in text: text = text.replace("?\"","\"?")
    text = text.replace(".",".<stop>")
    text = text.replace("?","?<stop>")
    text = text.replace("!","!<stop>")
    text = text.replace("<prd>",".")
    sentences = text.split("<stop>")
    sentences = sentences[:-1]
    sentences = [s.strip() for s in sentences]

    return sentences
```

## Function to extract entities from a sentence

```python
def get_entities(sent):
  """
  Extracts entities from a sentence

  :param text (str): single string of text from document
  :return [ent1.strip(), ent2.strip()] (tuple): tuple with entities in sentences
  """
  ## chunk 1
  ent1 = ""
  ent2 = ""

  prv_tok_dep = ""    # dependency tag of previous token in the sentence
  prv_tok_text = ""   # previous token in the sentence

  prefix = ""
  modifier = ""

  #############################################################
  
  for tok in nlp(sent):
    ## chunk 2
    # if token is a punctuation mark then move on to the next token
    if tok.dep_ != "punct":
      # check: token is a compound word or not
      if tok.dep_ == "compound":
        prefix = tok.text
        # if the previous word was also a 'compound' then add the current word to it
        if prv_tok_dep == "compound":
          prefix = prv_tok_text + " "+ tok.text
      
      # check: token is a modifier or not
      if tok.dep_.endswith("mod") == True:
        modifier = tok.text
        # if the previous word was also a 'compound' then add the current word to it
        if prv_tok_dep == "compound":
          modifier = prv_tok_text + " "+ tok.text
      
      ## chunk 3
      if tok.dep_.find("subj") == True:
        ent1 = modifier +" "+ prefix + " "+ tok.text
        prefix = ""
        modifier = ""
        prv_tok_dep = ""
        prv_tok_text = ""      

      ## chunk 4
      if tok.dep_.find("obj") == True:
        ent2 = modifier +" "+ prefix +" "+ tok.text
        
      ## chunk 5  
      # update variables
      prv_tok_dep = tok.dep_
      prv_tok_text = tok.text
  #############################################################

  return [ent1.strip(), ent2.strip()]
```

## Function to extract a relation from a sentence

```python
def get_relation(sent):
  """
  Extracts relations from a sentence

  :param text (str): single string of text from document
  :return (span.text) (): relation from sentence
  """
  doc = nlp(sent)

  # Matcher class object 
  matcher = Matcher(nlp.vocab)

  #define the pattern 
  pattern = [{'DEP':'ROOT'}, 
            {'DEP':'prep','OP':"?"},
            {'DEP':'agent','OP':"?"},  
            {'POS':'ADJ','OP':"?"}] 

  # matcher.add("matching_1", None, pattern)
  matcher.add("matching_1", [pattern])

  matches = matcher(doc)
  k = len(matches) - 1

  span = doc[matches[k][1]:matches[k][2]] 

  return (span.text)
```

In this notebook, the Instagram Terms and Services agreement will be used, which will be extracted from Instagram's website using the Python module Beautiful Soup.

### Pull text data from online source

```python
# att_url = "https://www.att.com/legal/terms.consumerServiceAgreement.html"
# apple_url = "https://www.apple.com/legal/sales-support/applecare/appmacnaen.html"
insta_url = "https://help.instagram.com/581066165581870"
html = urlopen(insta_url).read()
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()
```

## Preprocess text in document by:
- Removing blank sentences
- Stripping new line characters
- Splitting document into list of individual sentences
- Removing those sentences (through reading inspection and observation) that are not needed in knowledge graph because they do not provide important information to user

After preprocessing the text in the document and creating a list of sentences in the document that are preprocessed, the resultant sentences are saved in a dataframe.

```python
# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())

# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)
text = text.replace('\n', ' ')
sentences = split_into_sentences(text)
sentences = [sent for sent in sentences if len(sent.split())>3]
sentences = sentences[1:]

# remove sentences from dataset that are not necessary for knowledge graph
sent_to_remove = [
    'If you do not want to agree to these or any updated Terms, you can delete your account, here.',
    'Revised: 26 July 2022 Was this information helpful?',
    'YesNoRelated ArticlesWho can use Facebook Dating?',
    'Summary of updates to Meta Privacy Policy and Terms of ServiceWho can use Meta Pay?',
    'What happens to my digital collectibles if I delete or temporarily disable my Facebook account?',
    'How can I use lists to share to a specific group of people?',
    ', Menlo Park, CA 94025.',
    'Menlo Park, CA 94025.'
]
sentences = [sent for sent in sentences if sent not in sent_to_remove]

# create dataframe with sentences to be used in model development in knowledge graph
df = pd.DataFrame(columns=['sentences'], data=sentences)
```

## Extract entity pairs in sentences
Using a pre-trained model from the Spacy library, a part-of-speech tag, as well as dependency parser tags, are applied to each token from every sentence in the Instagram Terms and Services agreement document. Using the tags, the subjects and objects are identified from each sentence and will eventually be used in the entity-pair relations in the final knowledge graph.

By using dependency parser tags applied from Spacy's pre-trained model, more advanced grammatical labels are applied to each token to best determine entity-pairs. Further, because multiple entities can span across each sentence, the dependency parser tagging can help to alleviate the problem by focusing on more specific types of tags.

## __The 6-Ds: DOMAIN EXPERTISE__

__By using a pre-trained model for part-of-speech and dependency parser tagging words in a sentence, that has been developed by subject matter experts in linguistics, we gain domain expertise in natural language processing methods. For further model development, incorporating domain expertise from attorneys that develop legal agreements or contracts, improved performance in such methods as dependency parser tagging can create more robust knowledge graphs.__

```python
entity_pairs = []

for i in tqdm(df["sentences"]):
  entity_pairs.append(get_entities(i))
```

## Extract relations from sentences in document

```python
relations = [get_relation(i) for i in tqdm(df['sentences'])]
```

## View relations extracted from sentences; these relations will be used to focus on those entity-relation pairs based on user selection

```python
pd.Series(relations).value_counts()[:20]
```

## With entity-pair relations previously extracted, create dataframe" with:<br>source -> relation -> target pairs
The source will be the main node, the relation will represent the edge between the two nodes, and the target will represent node that the source node is directed to.

## __The 6-Ds: DESIGN__

__The design of this knowledge graph system involves an unsupervised ML approach. In the previous code cells, part-of-speech and dependency parser tagging was used to extract subject and objects, and finally entity-pair relations, to be used for nodes and edges in the resultant knowledge graph.__

```python
# extract subject
source = [i[0] for i in entity_pairs]
# extract object
target = [i[1] for i in entity_pairs]
kg_df = pd.DataFrame({'source':source, 'target':target, 'edge':relations})
```

Replace empty strings in dataframe created in previous code cell with NaN and remove rows with NaN.

```python
kg_df = kg_df.replace(r'^\s*$', np.nan, regex=True)
kg_df = kg_df.dropna()
kg_df
```

## Create directed graph from source -> relation -> target dataframe

```python
# create a directed-graph from a dataframe
G = nx.from_pandas_edgelist(kg_df, "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())
```

## View directed graph (knowledge graph)

## __The 6-Ds: DECOMPOSITION__
__A legal agreement or contract can contain several pages of text, and sometimes consisting of legal jargon or lexicon that can be difficult to comprehend. Any legal agreement or contract that a person enters into can affect an individual in many different ways, either requiring the user to abide by specific terms and agreements, or by a user benefitting or being supplied various types of products or services. By providing a tool such as a knowledge graph that can graphically models relationships among key entities in a legal agreement, and allows information retrieval about key events or topics, a user can more quickly and efficiently understand a legal agreement and reinvent the way a user enters into an agreement.__

```python
# plt.figure(figsize=(20, 20))
plt.rcParams["figure.figsize"] = (20, 20)
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color='red', edge_cmap=plt.cm.Blues, pos = pos)
# plt.show()
plt.savefig('instagram_kg.png')
```

## From list of relations, select word of interest to focus on only those entity-pair relations
In the cell below, we focus on the word "use" and only those entity-pair relations that have the relation "use"

```python
G=nx.from_pandas_edgelist(kg_df[kg_df['edge']=="use"], "source", "target", 
                          edge_attr=True, create_using=nx.MultiDiGraph())

plt.figure(figsize=(20,20))
pos = nx.spring_layout(G, k = 0.5) # k regulates the distance between nodes
nx.draw(G, with_labels=True, node_color='red', node_size=1500, edge_cmap=plt.cm.Blues, pos = pos)
# plt.show()
plt.savefig('instagram_use_kg.png')
```

## Extract sources, targets, and edges from knowledge graph data frame developed from Instagram Terms and Services agreement to create dictionary to be used in Cytoscape package for Dash web app UI

```python
sources = kg_df['source'].to_list()
targets = kg_df['target'].to_list()
edges = kg_df['edge'].to_list()
```

### Extract nodes labels to apply to nodes in Cytoscape knowledge graph

```python
node_labels = set(sources+targets)
```

### Apply node indices to each source label that will be used in dictionary for Cytoscape module

```python
from num2words import num2words
label_indices = {}
for label in enumerate(node_labels):
    label_indices[label[1]] = num2words(label[0])
```

```python
ids = []
for word in sources:
    ids.append(label_indices[word])
```

```python
kg_df['source_id'] = ids
```

### Apply node indices to each target label that will be used in dictionary for Cytoscape module

```python
ids = []
for word in targets:
    ids.append(label_indices[word])
```

```python
kg_df['target_id'] = ids
```

```python
from num2words import num2words
node_ids = {}
for label in enumerate(node_labels):
    node_ids[num2words(label[0])] = label[1]
```

### Create node elements dictionary to be used for Cytoscape module

```python
elements = []
x = 0
y = 0

for key, value in node_ids.items():
    temp = {}
    temp['data'] = {}
    temp['position'] = {}
    temp['position']['x'] = x
    temp['position']['y'] = y
    x+=25
    y+=25
    temp['data']['id'] = key
    temp['data']['label'] = value
    elements.append(temp)
```

### Create edge elements dictionary to be used for Cytoscape module

```python
edges = []
for source_id, target_id, edge in zip(kg_df['source_id'].to_list(), kg_df['target_id'].to_list(), kg_df['edge'].to_list()):
    temp = {}
    temp['data'] = {}
    temp['data']['source'] = source_id
    temp['data']['target'] = target_id
    temp['data']['label'] = edge
    edges.append(temp)
```

### Combine node elements and edge elements into single elements list of dictionaries to be used in Cytoscape module

```python
elements = elements+edges
```
